```python
import os
os.environ["KMP_DUPLICATE_LIB_OK"]="TRUE"
%matplotlib inline
import numpy as np
import torch
import matplotlib.pyplot as plt
#torch.set_printoptions(edgeitems=2, linewidth=75)
```

# Define inputs and outputs for the model

```python
t_c = torch.tensor([0.5,  14.0, 15.0, 28.0, 11.0,  8.0,  3.0, -4.0,  6.0, 13.0, 21.0])
t_u = torch.tensor([35.7, 55.9, 58.2, 81.9, 56.3, 48.9, 33.9, 21.8, 48.4, 60.4, 68.4])
```

<p style="font-size:20px">Model</p>
<p style="text-align:center">$\huge m = w \cdot t_u + b$</p>
<p style="text-align:center">$\huge \frac {\partial m}{\partial w} = t_u $</p>
<p style="text-align:center">$\huge \frac {\partial m}{\partial b} = 1 $</p>


```python
def model(t_u, w, b):
    return t_u*w + b

def dmodel_dw(t_u, w, b):
    return t_u

def dmodel_db(t_u, w, b):
    return 1
```

<p style="font-size:20px">Loss</p>
<p style="text-align:center">$\huge loss(t_u, t_c) = [m(t_u) - t_c]^2$</p>
<p style="text-align:center">$\huge \frac {\partial loss(t_u, t_c)}{\partial m} = 2 \cdot [m(t_u) - t_c] $</p>

```python
def loss_fn(t_p, t_c):
    return torch.mean(torch.square(t_p-t_c))

def dloss_fn_dm(t_p, t_c):
    return (2*(t_p-t_c)) / t_p.shape[0] # 1/N is from the mean. 
```

<p style="font-size:20px">Gradient</p>
<p style="text-align:center">$\huge \nabla _{w,b}L(w,b, t_u, t_c) = (\frac{\partial L}{\partial w},\frac{\partial L}{\partial b})$
$\huge =(\frac{\partial L}{\partial m}\cdot\frac{\partial m}{\partial w}, \frac{\partial L}{\partial m}\cdot\frac{\partial m}{\partial b})$
</p>

```python
def grad_fn(t_u, t_c, t_p, w, b):
    dloss_dm = dloss_fn_dm(t_p, t_c)
    dloss_dw = dloss_dm * dmodel_dw(t_u, w, b)
    dloss_db = dloss_dm * dmodel_db(t_u, w, b)
    return torch.stack([dloss_dw.sum(), dloss_db.sum()])
```

# The Training Loop
<p style="text-align:center">$\huge w_{t+1} = w_t - lr \cdot \nabla _{w}L $</p>
<p style="text-align:center">$\huge b_{t+1} = b_t - lr \cdot \nabla _{b}L $</p>

```python
def training_loop(n_epochs, learning_rate, params, t_u, t_c):
    for epoch in range(1, n_epochs + 1):
        w, b = params
        t_p = model(t_u, w, b) # forward pass: compute the model prediction
        loss = loss_fn(t_p, t_c) # computre the loss (to monitor conversion)
        grad = grad_fn(t_u, t_c, t_p, w, b) #  backward pass: use local loss gradient to
                                            # compute how loss responds to parameter changes
        params = params - learning_rate * grad # update parameters
        if epoch % 10 ==0:
            print('Epoch %d, Loss %f' % (epoch, float(loss)))
    return params
```

# Run the model

```python
training_loop(
        n_epochs = 100, 
        learning_rate = 1e-4,
        params = torch.tensor([1.0, 0.0]), # (w_0, b_0)
        t_u = t_u,
        t_c = t_c)
```

## Plot convergence

```python
def training_loop(n_epochs, learning_rate, params, t_u, t_c):
    log = []
    for epoch in range(1, n_epochs + 1):
        w, b = params
        t_p = model(t_u, w, b) # forward pass: compute the model prediction
        loss = loss_fn(t_p, t_c) # computre the loss (to monitor conversion)
        grad = grad_fn(t_u, t_c, t_p, w, b) #  backward pass: use local loss gradient to
                                            # compute how loss responds to parameter changes
        params = params - learning_rate * grad # update parameters

        log.append( torch.tensor([epoch, loss, params[0], params[1], grad[0], grad[1] ]) ) # epoch, loss, w, b, grad_w, grad_b
    return params, torch.stack(log)
```

```python
params, log = training_loop(
        n_epochs = 100, 
        learning_rate = 1e-4,
        params = torch.tensor([1.0, 0.0]),
        t_u = t_u,
        t_c = t_c)
print(f"w={params[0]:3.3f}, b={params[1]:3.3f}")
```

```python
fig, ax = plt.subplots(figsize=(6,4))
ax.plot(log[:,0], log[:,1])
ax.set_xlabel('epoch', fontsize=20);
ax.set_ylabel('loss', fontsize=20);
```

```python
def plot_progress(log):
    fig, ax = plt.subplots(figsize=(6,4))
    ax.plot(log[:,0], log[:,1])
    ax.set_xlabel('epoch', fontsize=20);
    ax.set_ylabel('loss', fontsize=20);

    fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12,9))
    axes[0,0].plot(log[:,0], log[:,2])
    axes[0,0].set_xlabel('epoch', fontsize=20);
    axes[0,0].set_ylabel('w', fontsize=20);

    axes[0,1].plot(log[:,0], log[:,3])
    axes[0,1].set_xlabel('epoch', fontsize=20);
    axes[0,1].set_ylabel('b', fontsize=20);

    axes[1,0].plot(log[:,0], log[:,4])
    axes[1,0].set_xlabel('epoch', fontsize=20);
    axes[1,0].set_ylabel(r'$\nabla _{w}L$', fontsize=20);

    axes[1,1].plot(log[:,0], log[:,5])
    axes[1,1].set_xlabel('epoch', fontsize=20);
    axes[1,1].set_ylabel(r'$\nabla _{b}L$', fontsize=20);
```

```python
plot_progress(log)
```

Notes: 

- The procedure is MUCH faster than complete enumeration. 
- It did not converge to the correct values (0.5556, -17.78). The loss function is very high!

Why? 
- The training loop performance is very sensitive to the learning rate. With proper learning rate, the parameter updates get very small. So, we need adaptive $lr$. 
- Another problem: The extremely different values of the gradients for $w$ and $w$: $\nabla _{w}L >> \nabla _{b}L$<br>
This suggests that each parameter must have a different $lr$. 

Solutions:
- Separate $lr$ for each parameter or parameter normalization. 
- Alternatively, normalize the inputs. 
- Perhaps, the model did not yet converge. Increase the number of iterations. 

# Normalizing inputs
$\Large t_u \Rightarrow \frac {t_u-mean(t_u)}{std(t_u)} $
Note faster convergence, even with higher learning rate. 

```python
params, log = training_loop(
        n_epochs = 100, 
        learning_rate = 1e-2,
        params = torch.tensor([1.0, 0.0]),
        t_u = 0.1*t_u, # or (t_u-t_u.mean())/t_u.std(),
        t_c = t_c)
print(f"w={params[0]:3.3f}, b={params[1]:3.3f}")
plot_progress(log)
```

```python
params, log = training_loop(
        n_epochs = 5000, 
        learning_rate = 1e-2,
        params = torch.tensor([1.0, 0.0]),
        t_u = 0.1*t_u, #(t_u-t_u.mean())/t_u.std(),
        t_c = t_c)
print(f"w={params[0]:3.3f}, b={params[1]:3.3f}")
plot_progress(log)
```

```python
params, log1 = training_loop(
        n_epochs = 5000, 
        learning_rate = 1e-2,
        params = torch.tensor([1.0, 0.0]),
        t_u = (t_u-t_u.mean())/t_u.std(),
        t_c = t_c)
print(f"w={params[0]:3.3f}, b={params[1]:3.3f}")
plot_progress(log)
```

```python
params, log2 = training_loop(
        n_epochs = 5000, 
        learning_rate = 1e-4,
        params = torch.tensor([1.0, 0.0]),
        t_u =t_u, #(t_u-t_u.mean())/t_u.std(),
        t_c = t_c)
print(f"w={params[0]:3.3f}, b={params[1]:3.3f}")
plot_progress(log)
```
