```python
import torch

a = torch.tensor(-1.0, requires_grad=True)
b = torch.tensor(3.0, requires_grad=True)
c = torch.tensor(5.0, requires_grad=True)

def eq(x, a, b, c):
    return a**2 * x + torch.cos(b * x) + c * x

y = eq(0.76, a, b, c)
y.backward()

print(a.grad, b.grad, c.grad)
```

```python
# manual
import torch

a = torch.tensor(-1.0, requires_grad=True)
b = torch.tensor(3.0, requires_grad=True)
c = torch.tensor(5.0, requires_grad=True)

def eq(x, a, b, c):
    return a**2 * x + torch.cos(b * x) + c * x

x = 0.76
y = eq(x, a, b, c)

print("y =", y.item())

y.backward()

print("dy/da =", a.grad.item())
print("dy/db =", b.grad.item())
print("dy/dc =", c.grad.item())
```

```python
# Manual derivative calculation for:
# y = a**2 * x + cos(b*x) + c*x

import torch

# same values as in the previous cell
x = 0.76
a_val = a.detach().item()
b_val = b.detach().item()
c_val = c.detach().item()

# manual gradients from the formulas:
# dy/da = 2 * a * x
# dy/db = -x * sin(b * x)
# dy/dc = x

dy_da_manual = 2 * a_val * x
dy_db_manual = -x * torch.sin(torch.tensor(b_val * x)).item()
dy_dc_manual = x

print("Manual gradients:")
print("dy/da =", dy_da_manual)
print("dy/db =", dy_db_manual)
print("dy/dc =", dy_dc_manual)

print("\nPyTorch gradients:")
print("dy/da =", a.grad.item())
print("dy/db =", b.grad.item())
print("dy/dc =", c.grad.item())
```
